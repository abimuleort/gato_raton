```python
import random
import copy 
import math
import time
import os

n = 10
raton = "üêÄ"
gato = "üêà‚Äç‚¨õ"
simbolo_queso = "üßÄ"
simbolo_vacio = "."

pos_raton = (2, 3)
pos_gato = (5, 6)
pos_queso = (8, 1)

def crear_tablero(n):
    tablero = []
    for _ in range(n):
        fila = [simbolo_vacio] * n
        tablero.append(fila)
    return tablero

def imprimir_tablero(tablero, gato, raton, queso):
    os.system("cls" if os.name == "nt" else "clear")
    for i in range(len(tablero)):
        fila = " "
        for j in range(len(tablero)):
            if (i, j) == gato:
                fila += simbolo_gato + " "
            elif (i, j) == raton:
                fila += simbolo_raton + " "
            else: 
                fila += simbolo_vacio + " "
        print(fila)
    print()

def movimientos_validos(pos, n):
    x, y = pos
    dirs = [(-1,0), (1,0), (0,-1), (0,1)]
    validos = []
    for dx, dy in dirs:
        nx, ny = x + dx, y + dy
        if 0 <= nx < n and 0 <= ny < n:
            validos.append((nx, ny))
        return validos
    
def distancia_manhattan(a, b):
   return abs(a[0]-b[0]) + abs(a[1]-b[1])
    
def valor_heuristico(gato, raton):
    d = distancia_manhattan(gato, raton)
    if gato == raton:
        return -math.inf
    return d

def minimax(gato, raton, profundidad, es_turno_raton, alfa, beta, n):
    if profundidad == 0 or gato == raton:
        return valor_heuristico(gato, raton)
    
    if es_turno_raton:
        mejor = -math.inf
        for mov in movimientos_validos(raton, n):
            valor = minimax(gato, mov, profundidad -1, False)
            mejor = max(mejor, valor)
            alfa = max(alfa, mejor)
            if beta <= alfa:
                break
        return mejor
    
    else: 
        mejor = math.inf
        for mov in movimientos_validos(gato, n):
            valor = minimax(mov, raton, profundidad -1, True, n)
            mejor = min(mejor, valor)
            beta = min(beta, mejor)
            if beta <= alfa:
                break
        return mejor

tablero = crear_tablero(n)
imprimir_tablero(tablero, pos_gato, pos_raton, pos_queso)

# condiciones de victoria

if pos_gato == pos_raton:
    print ("¬°El gato atrapo al raton! Fin del juego.")

if pos_raton == pos_queso:
    print("¬°El rat√≥n lleg√≥ al queso!")


mov = input("Movimiento w: arriba, s: abajo, a: derecha, d: izquierda")
def mover(pos, mov):
  x, y = simbolo_raton
if mov == "w": nueva_pos = (x -1, y) 
elif mov == "s": nueva_pos = (x +1, y)
elif mov == "a": nueva_pos = (x, y -1)
elif mov == "d": nueva_pos = (x, y +1)

```
