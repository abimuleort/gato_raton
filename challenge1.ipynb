```python
import random # importo el m√≥dulo para generar numeros aleatorios (decidir qui√©n empieza)
import copy   # importo el m√≥dulo para que el minimax pueda probar jugadas hipot√©ticas (no afecta el estado actual del juego)
import math   # importo funciones matem√°ticas para calcular todos los valores posibles de los movimientos
import os     # interactua con el sistema operativo para limpiar la consola
import time   # crea pausas, simula pensamiento de la computadora durante turnos 

# ------------- Configuracion Inicial -------------
n = 10 # establezco el tama√±o del tablero de juego (10x10)

       # defino los simbolos o representaciones
raton = "üêÄ"
gato = "üêà‚Äç‚¨õ"
queso = "üßÄ"
vacio = "."

# ------------- Posiciones iniciales -------------
# indico las posiciones de cada elemento en el tablero (fila, columna)
pos_raton = (2, 3)
pos_gato = (5, 6)
pos_queso = (8, 1)

# ------------- Creacion del tablero -------------
def crear_tablero(n):           # defino una funcion que recube el tama√±o del tablero
    tablero = []                # creo una lista vacia del tablero
    for _ in range(n):          #se crea un bucle donde se de un numero n 
        fila = [vacio] * n      #cada fila se multiplica por un . que se multiplica por un numero a definir 
        tablero.append(fila)    #y de eso se imprime en el tablero nuevas filas 
    return tablero              #con esto se resume tu tablero 

def imprimir_tablero(tablero, pos_gato, pos_raton, pos_queso):
    os.system("cls" if os.name == "nt" else "clear")
    for fila in range(len(tablero)):
        linea = ""
        for col in range(len(tablero)):
            if (fila, col) == pos_gato:
                linea += gato + " "
            elif (fila, col) == pos_raton:
                linea += raton + " "
            elif (fila, col) == pos_queso:
                linea += queso + " "
            else:
                linea += vacio + " "
        print(linea)
print()

# ------------- Programa principal --------------
tablero = crear_tablero(n)
imprimir_tablero(tablero, pos_gato, pos_raton, pos_queso) 

# ------------- Condiciones del juego -------------
```
