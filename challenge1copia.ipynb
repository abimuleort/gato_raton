```python
import random
import math
import copy
import os
import time

# CONFIGURACIN INICIAL
TAMANIO = 5
MAX_PROFUNDIDAD = 3  # Profundidad del 谩rbol de b煤squeda

VACIO = "."
GATO = "G"
RATON = "R"

# FUNCIONES DE UTILIDAD
def crear_tablero(n):
    return [[VACIO for _ in range(n)] for _ in range(n)]

def imprimir_tablero(tablero, gato, raton):
    os.system("cls" if os.name == "nt" else "clear")
    for i in range(len(tablero)):
        fila = ""
        for j in range(len(tablero)):
            if (i, j) == gato:
                fila += GATO + " "
            elif (i, j) == raton:
                fila += RATON + " "
            else:
                fila += VACIO + " "
        print(fila)
    print()

def movimientos_validos(pos, n):
    x, y = pos
    dirs = [(-1,0), (1,0), (0,-1), (0,1)]  # 4 direcciones
    validos = []
    for dx, dy in dirs:
        nx, ny = x + dx, y + dy
        if 0 <= nx < n and 0 <= ny < n:
            validos.append((nx, ny))
    return validos

def distancia_manhattan(a, b):
    return abs(a[0]-b[0]) + abs(a[1]-b[1])


# FUNCIN HEURSTICA
def valor_heuristico(gato, raton):
    d = distancia_manhattan(gato, raton)
    if gato == raton:
        return -math.inf  # gato gana
    return d  # m谩s distancia => mejor para el rat贸n


# IMPLEMENTACIN MINIMAX CON PODA ALFA-BETA
def minimax(gato, raton, profundidad, es_turno_raton, alfa, beta):
    if profundidad == 0 or gato == raton:
        return valor_heuristico(gato, raton)

    if es_turno_raton:
        mejor = -math.inf
        for mov in movimientos_validos(raton, TAMANIO):
            valor = minimax(gato, mov, profundidad - 1, False, alfa, beta)
            mejor = max(mejor, valor)
            alfa = max(alfa, mejor)
            if beta <= alfa:
                break
        return mejor
    else:
        mejor = math.inf
        for mov in movimientos_validos(gato, TAMANIO):
            valor = minimax(mov, raton, profundidad - 1, True, alfa, beta)
            mejor = min(mejor, valor)
            beta = min(beta, mejor)
            if beta <= alfa:
                break
        return mejor

# -----------------------------
# MOVIMIENTOS INTELIGENTES
# -----------------------------
def mejor_movimiento_raton(gato, raton):
    mejor_valor = -math.inf
    mejor_mov = raton
    for mov in movimientos_validos(raton, TAMANIO):
        valor = minimax(gato, mov, MAX_PROFUNDIDAD, False, -math.inf, math.inf)
        if valor > mejor_valor:
            mejor_valor = valor
            mejor_mov = mov
    return mejor_mov

def mejor_movimiento_gato(gato, raton):
    mejor_valor = math.inf
    mejor_mov = gato
    for mov in movimientos_validos(gato, TAMANIO):
        valor = minimax(mov, raton, MAX_PROFUNDIDAD, True, -math.inf, math.inf)
        if valor < mejor_valor:
            mejor_valor = valor
            mejor_mov = mov
    return mejor_mov

# -----------------------------
# SIMULACIN PRINCIPAL
# -----------------------------
def juego():
    gato = (0, 0)
    raton = (TAMANIO-1, TAMANIO-1)
    tablero = crear_tablero(TAMANIO)
    turnos = 0

    while True:
        imprimir_tablero(tablero, gato, raton)
        print(f"Turno: {turnos}")
        time.sleep(0.8)

        # Turno del rat贸n
        raton = mejor_movimiento_raton(gato, raton)
        imprimir_tablero(tablero, gato, raton)
        time.sleep(0.8)

        if gato == raton:
            print(" El gato atrap贸 al rat贸n.")
            break

        turnos += 1
        if turnos >= 25:
            print(" El rat贸n logr贸 sobrevivir, escap贸 del laberinto.")
            break

# -----------------------------
# EJECUCIN
# -----------------------------
if __name__ == "__main__":
    juego()
```
