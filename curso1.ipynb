```python
saldo = 1000
print("\t.: MENU:.")
print("1. ingresar dinero a la cuenta")
print("2. retirar dinero en la cuenta")
print("3. mostrar dinero disponible")
print("4. salir")
opcion = int(input("digite una opcion de menu"))

print()

if opcion==1:
    extra = float(input("cuanto dinero desea ingresar"))
    saldo += extra
    print(f"dinero en la cuenta: {saldo}")
elif opcion==2:
    retirar = float(input("cuanto dinero desea retirar"))
    if retirar>saldo:
        print("no tiene saldo disponible")
    else:
        saldo -= retirar 
        print(f"dinero de la cuenta: {saldo}")
elif opcion==3:
    print(f"dinero en la cuenta: {saldo}")
elif opcion==4:
    print("gracias por usar el cajero automatico")
else:
    print("ha digitado una opcion incorrecta")
```

```python
# listas
lista = ['lunes', 'martes', 'miercoles', 'jueves', 'viernes']
print(lista[3], lista[-1]) # imprime elementos segun su indice
print(lista[0:3]) # imprime sublistas 
print(lista[2:])  # para imprimir desde un elemento hasta el final se deja vacio despues de los dos puntos
print(len(lista)) # len para imprimir la cantidad de elementos que tiene la lista
```

```python
# lista pt2
lista = [1, 1, 2, 2, 2, 3, 4, 5, "abi", "abi"]
lista.append(6)        # append para agregar elementos al final de la lista
# lista.insert(2, 3)   # insert para agregar elemento en un lugar especifico a la lista
# lista.extend([7, 8]) # extend para agregar varios elementos a una lista de forma rapida
print(3 in lista)      # 3 in lista (in) para preguntar si un elemento se encuentra dentro de la lista
print(3 in lista)
print(lista.index(5))  # index para buscar en que indice se encuentra cierto elemento
print(lista.count(1))  # count para saber cuantas veces se repite un elemento en la lista
lista.pop()            # si se deja vacio el parentesis se elimina el ultimo elemento de la lista, si se quiere eliminar un elemento determinado se indica el indice
print(lista)
lista.remove("abi")    # para eliminar elementos de lista sin indicar indice, sino contenido
print(lista)
lista.reverse()        # voltea todo el orden de la lista
print(lista)
listaa = [2, 1, 5, 3, 4, 6, 9, 8, 7]
listaa.sort()          # ordena los elementos de la lista, si se desea ordenar de forma descendente se ingresa en el parentesis "reverse.True"
print(listaa)
```

```python
# tuplas: son listas inmutables
tupla = (4, "HOla", 5.40, [1, 2, 3], 4)
# las tupls no se pueden modificar como las listas pero si se pueden hacer busquedas en ellas
lista = list(tupla) # con el metodo list se convierte una tupla en lista, sus elementos no se modifican, y se puede realizar lo mismo y viceversa
tupla = tuple(lista)
print(tupla)
print(lista)
```
